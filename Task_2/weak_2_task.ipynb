# Question : 1

```python
def ip_address(string):
    if "." in string:
        valid_ip = ""
        for i in string:
            if i == ".":
                valid_ip += "[.]"
            else:
                valid_ip += i
        return valid_ip
    else:
        return string
```

```python
ip_add = "123.noty.no.n.newly.new."
valid_ip_address = ip_address(ip_add)
print(f"Valid IP Address : {valid_ip_address}")
```

# Question : 2

```python
def even_digit(digit):
    count = 0
    for i in digit:
        if len(str(i))%2 == 0:
            count +=1
    return count
```

```python
listA = [11,425,64,634,538534]
count = even_digit(listA)
print(f"There are {count} even number of digits.")
```

# Question : 3

```python
def good_pair(list_):
    count = 0
    for i in range(len(list_)):
        for j in range(i+1,len(list_)):
            if list_[i] == list_[j]:
                count += 1
    return count
```

```python
list_ = [1,2,3,1,1,3]
count = good_pair(list_)
print(f"Number of good pairs : {count}")
```

# Question : 4

```python
def smaller_count(list_):
    count_list = []
    for i in range(len(list_)):
        count = 0
        for j in range(len(list_)):
            if list_[i] > list_[j]:
                count += 1
        count_list.append(count)
    return count_list
```

```python
list_ = [6,5,4,8]
count_list_ = smaller_count(list_)
print(f"List of numbers are the smaller than their Current number : {count_list_}")
```

# Question : 5

```python
def diff_num(num):
    summition = 0
    product = 1
    for i in str(num):
        summition += int(i)
        product *= int(i)
    return product - summition
```

```python
n = 4421
diff = diff_num(n)
print(f"Difference between the product and the sum of its digits : {diff}")
```

# Question : 6

```python
def xor_operation(n, start):
    ans = 0
    for i in range(n):
        num = start + 2 * i
        ans ^= num
    return ans
```

```python
n = 4
start = 3
opr = xor_operation(n, start)
print(f"Bitwise XOR of all elements : {opr}")
```
